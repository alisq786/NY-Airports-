```python
# Average Departure and Arrival Delays by Origion
n_groups = 3
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .85
 
rects1 = plt.bar(index, avg_delay['dep_delay'], bar_width,
                 alpha=opacity,
                 color='blue',
                 label='Departure')
 
rects2 = plt.bar(index + bar_width, avg_delay['arr_delay'], bar_width,
                 alpha=opacity,
                 color='green',
                 label='Arrival')
 
plt.xlabel('Origin')
plt.ylabel('Average Delay (Minutes)')
plt.title('Average Delay by Origin')
mob_vs_web_by_agegrp = plt.xticks(index + bar_width, ('EWR', 'JFK', 'LGA'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

```python

def depart_ampm(row):
    if row['sched_dep_time'] >= 0 and row['sched_dep_time'] < 1200:
        return 'AM'
    elif row['sched_dep_time'] >= 1200 and row['sched_dep_time'] < 2400:
        return 'PM'
   

data_airline['depart_time'] = data_airline.apply(depart_ampm,axis=1)

data_airline.tail()
```

```python
#delay_DN=data_airline.groupby(['depart_time'])[['dep_delay','arr_delay']].mean()
delay_dep_DN=data_airline.groupby(['depart_time'])[['dep_delay']].mean()
delay_dep_DN
```

```python
# simpler way to do
#delay_DN.plot(delay_DN.index, ['dep_delay','arr_delay'], kind='bar', color=['b','g'])
delay_dep_DN.plot(delay_DN.index, ['dep_delay'], kind='bar', color=['b','g'])
plt.ylabel('Average Delay (Minutes)')
plt.title('Average Flight Delay By Departure Time (AM/PM)')
```

```python
def arr_ampm(row):
    if row['sched_arr_time'] >= 0 and row['sched_arr_time'] < 1200:
        return 'AM'
    elif row['sched_arr_time'] >= 1200 and row['sched_arr_time'] < 2400:
        return 'PM'
   

data_airline['arrival_time'] = data_airline.apply(arr_ampm,axis=1)

data_airline.tail()
```

```python
delay_arr_DN=data_airline.groupby(['arrival_time'])[['arr_delay']].mean()
delay_arr_DN
```

```python
delay_arr_DN.plot(delay_arr_DN.index, ['arr_delay'], kind='bar', color=['b','g'])
plt.ylabel('Average Delay (Minutes)')
plt.title('Average Flight Delay By Arrival Time (AM/PM)')
```
